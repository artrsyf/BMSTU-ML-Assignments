```python
%pip install openpyxl
import io
import requests
import pandas as pd

url = "https://www.worldbank.org/content/dam/sites/govindicators/doc/wgidataset_with_sourcedata-2025.xlsx"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "*/*",
    "Referer": "https://www.worldbank.org/"
}

response = requests.get(url, headers=headers, timeout=60)
response.raise_for_status()

excel_file = io.BytesIO(response.content)

dfs = pd.read_excel(excel_file, engine="openpyxl", sheet_name=None)
df = pd.concat(dfs.values(), ignore_index=True)

df.head()
```

```python
import pandas as pd
from IPython.display import display
import ipywidgets as widgets

target_year = 2024

# Предположим, что df уже загружен

# Получаем все уникальные Governance dimension
dimensions = df_selected["Governance dimension"].unique()

# Создаём Dropdown для выбора измерения
dimension_dropdown = widgets.Dropdown(
    options=dimensions,
    description='Dimension:',
    value=dimensions[0],
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

# Виджет для выбора года
year_dropdown = widgets.Dropdown(
    options=sorted(df_selected["Year"].unique(), reverse=True),
    description='Year:',
    value=target_year,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='200px')
)

# Функция обновления таблицы при выборе
def update_table(change):
    selected_dim = dimension_dropdown.value
    selected_year = year_dropdown.value
    df_filtered = df_selected[
        (df_selected["Year"] == selected_year) &
        (df_selected["Governance dimension"] == selected_dim)
    ].sort_values(by="Governance score (0-100)", ascending=False)
    display(df_filtered.head(20))

# Используем interactive_output, чтобы обновлять таблицу автоматически
out = widgets.interactive_output(update_table, {'change': dimension_dropdown})

# Показываем виджеты и вывод
display(widgets.HBox([dimension_dropdown, year_dropdown]), out)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

dimension_dropdown = widgets.Dropdown(
    options=dimensions,
    description='Dimension:',
    value=dimensions[0],
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

year_dropdown = widgets.Dropdown(
    options=sorted(df_selected["Year"].unique(), reverse=True),
    description='Year:',
    value=2024,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='200px')
)

top_n_slider = widgets.IntSlider(
    value=20,
    min=5,
    max=100,
    step=5,
    description='Top N:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)

def plot_wgi(dimension, year, top_n):
    df_filtered = df_selected[
        (df_selected["Year"] == year) & 
        (df_selected["Governance dimension"] == dimension)
    ].sort_values(by="Governance score (0-100)", ascending=False).head(top_n)
    
    df_filtered = df_filtered[::-1]  # разворачиваем, чтобы 1 был сверху
    
    plt.figure(figsize=(10, max(5, len(df_filtered)/2)))
    
    bars = plt.barh(range(len(df_filtered)), df_filtered["Governance score (0-100)"], color='green')
    
    # Нумерация стран сверху вниз
    plt.yticks(range(len(df_filtered)), [f"{len(df_filtered)-i}. {c}" for i, c in enumerate(df_filtered["Economy (name)"])])
    
    ax = plt.gca()
    ax.set_ylim(-0.5, len(df_filtered)-0.5)
    
    # Создаём верхнюю ось с такой же шкалой
    ax_top = ax.twiny()
    ax_top.set_xlim(ax.get_xlim())
    ax_top.set_xticks(ax.get_xticks())
    ax_top.set_xticklabels([str(int(x)) for x in ax.get_xticks()])
    
    # Подписи значений справа от столбцов
    for i, v in enumerate(df_filtered["Governance score (0-100)"]):
        ax.text(v + 0.5, i, f"{v:.1f}", va='center', ha='left', fontsize=9, color='black')
    
    # Подписи и сетка
    ax.set_xlabel("Governance score (0-100)")
    plt.title(f"WGI {dimension} - {year}")
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    
    # Убираем лишние отступы
    plt.tight_layout()
    plt.margins(y=0)
    
    plt.show()

out = widgets.interactive_output(plot_wgi, {
    'dimension': dimension_dropdown,
    'year': year_dropdown,
    'top_n': top_n_slider
})

display(widgets.HBox([dimension_dropdown, year_dropdown, top_n_slider]), out)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Предположим, что df_filtered_region уже загружен
dimensions = df_filtered_region["Governance dimension"].unique()
years = sorted(df_filtered_region["Year"].unique(), reverse=True)

# Виджет для выбора метрики
dimension_dropdown = widgets.Dropdown(
    options=dimensions,
    description='Dimension:',
    value=dimensions[0],
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

# Виджет для выбора года для анализа max/min
year_dropdown = widgets.Dropdown(
    options=years,
    description='Year:',
    value=2024,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='200px')
)

# Функция для обновления данных
def update_wgi_stats(dimension, year):
    # Фильтруем по dimension и году
    df_dim = df_filtered_region[df_filtered_region["Governance dimension"] == dimension]
    
    # 1️⃣ Страны с наибольшим и наименьшим WGI за выбранный год
    df_2024 = df_dim[df_dim["Year"] == year][["Economy (name)", "Governance estimate (approx. -2.5 to +2.5)"]]
    max_value = df_2024["Governance estimate (approx. -2.5 to +2.5)"].max()
    min_value = df_2024["Governance estimate (approx. -2.5 to +2.5)"].min()
    max_countries = df_2024[df_2024["Governance estimate (approx. -2.5 to +2.5)"] == max_value]["Economy (name)"].tolist()
    min_countries = df_2024[df_2024["Governance estimate (approx. -2.5 to +2.5)"] == min_value]["Economy (name)"].tolist()
    
    print(f"Страны с наибольшим WGI за {year} (Estimate = {max_value:.2f}): {', '.join(max_countries)}")
    print(f"Страны с наименьшим WGI за {year} (Estimate = {min_value:.2f}): {', '.join(min_countries)}\n")
    
    # 2️⃣ Среднее значение региона по годам
    df_region_estimate = df_dim[["Year", "Governance estimate (approx. -2.5 to +2.5)"]]
    region_mean_per_year = df_region_estimate.groupby("Year").mean().reset_index()
    region_mean_per_year.rename(columns={"Governance estimate (approx. -2.5 to +2.5)":"Regional Mean Estimate"}, inplace=True)
    
    display(region_mean_per_year)

# Создаем интерактивный вывод
out = widgets.interactive_output(update_wgi_stats, {
    'dimension': dimension_dropdown,
    'year': year_dropdown
})

display(widgets.HBox([dimension_dropdown, year_dropdown]), out)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# df_filtered_region - страны региона Americas
# df_selected - полный датасет с Россией

dimensions = df_selected["Governance dimension"].unique()

dimension_dropdown = widgets.Dropdown(
    options=dimensions,
    description='Dimension:',
    value=dimensions[0],
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

def update_summary_region(dimension):
    # Фильтруем по выбранной метрике для региона
    df_region_score = df_filtered_region[df_filtered_region["Governance dimension"] == dimension][
        ["Economy (name)", "Region", "Year", "Governance score (0-100)"]
    ]
    
    # Добавляем Россию из полного датасета
    df_russia = df_selected[
        (df_selected["Economy (name)"] == "Russian Federation") &
        (df_selected["Governance dimension"] == dimension)
    ][["Economy (name)", "Region", "Year", "Governance score (0-100)"]]
    
    # Объединяем
    df_score = pd.concat([df_region_score, df_russia], ignore_index=True)
    
    # Выбираем 1996 и 2024 годы
    df_1996 = df_score[df_score["Year"] == 1996][["Economy (name)", "Region", "Governance score (0-100)"]].rename(
        columns={"Governance score (0-100)": "WGI 1996"})
    df_2024 = df_score[df_score["Year"] == 2024][["Economy (name)", "Region", "Governance score (0-100)"]].rename(
        columns={"Governance score (0-100)": "WGI 2024"})
    
    # Объединяем по странам
    df_combined = pd.merge(df_1996, df_2024, on=["Economy (name)", "Region"], how="outer")
    df_combined["Изменение (см. пункт 10)"] = df_combined["WGI 2024"] - df_combined["WGI 1996"]
    
    rows = []

    # mean_2024 по региону (без России)
    mean_1996 = df_region_score[df_region_score["Year"] == 1996]["Governance score (0-100)"].mean()
    mean_2024 = df_region_score[df_region_score["Year"] == 2024]["Governance score (0-100)"].mean()
    rows.append({
        "": "mean_2024",
        "Регион": "-",
        "Страна": "-",
        "WGI 1996": mean_1996,
        "WGI 2024": mean_2024,
        "Изменение (см. пункт 10)": mean_2024 - mean_1996
    })

    # max_2024 по региону (без России)
    max_val = df_region_score[df_region_score["Year"] == 2024]["Governance score (0-100)"].max()
    max_rows = df_region_score[
        (df_region_score["Year"] == 2024) &
        (df_region_score["Governance score (0-100)"] == max_val)
    ]
    for _, r in max_rows.iterrows():
        row_1996 = df_combined[df_combined["Economy (name)"] == r["Economy (name)"]]["WGI 1996"].values[0]
        rows.append({
            "": "max_2024",
            "Регион": r["Region"],
            "Страна": r["Economy (name)"],
            "WGI 1996": row_1996,
            "WGI 2024": r["Governance score (0-100)"],
            "Изменение (см. пункт 10)": r["Governance score (0-100)"] - row_1996
        })

    # min_2024 по региону (без России)
    min_val = df_region_score[df_region_score["Year"] == 2024]["Governance score (0-100)"].min()
    min_rows = df_region_score[
        (df_region_score["Year"] == 2024) &
        (df_region_score["Governance score (0-100)"] == min_val)
    ]
    for _, r in min_rows.iterrows():
        row_1996 = df_combined[df_combined["Economy (name)"] == r["Economy (name)"]]["WGI 1996"].values[0]
        rows.append({
            "": "min_2024",
            "Регион": r["Region"],
            "Страна": r["Economy (name)"],
            "WGI 1996": row_1996,
            "WGI 2024": r["Governance score (0-100)"],
            "Изменение (см. пункт 10)": r["Governance score (0-100)"] - row_1996
        })

    # Россия
    if not df_russia.empty:
        r = df_russia[df_russia["Year"] == 2024].iloc[0]
        row_1996 = df_combined[df_combined["Economy (name)"] == r["Economy (name)"]]["WGI 1996"].values[0]
        rows.append({
            "": "Russia",
            "Регион": r["Region"],
            "Страна": "Russia",
            "WGI 1996": row_1996,
            "WGI 2024": r["Governance score (0-100)"],
            "Изменение (см. пункт 10)": r["Governance score (0-100)"] - row_1996
        })
    
    df_summary = pd.DataFrame(rows)
    display(df_summary)

# Интерактивный виджет для региона + Россия
widgets.interactive(update_summary_region, dimension=dimension_dropdown)
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# df_selected должен содержать колонки:
# ["Economy (name)", "Region", "Year", "Governance estimate (approx. -2.5 to +2.5)", "Governance dimension"]

dimensions = df_selected["Governance dimension"].unique()

dimension_dropdown = widgets.Dropdown(
    options=dimensions,
    description='Dimension:',
    value=dimensions[0],
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

def plot_wgi_boxplot(dimension):
    # Фильтруем по выбранной метрике и по 2024 году
    df_plot = df_selected[
        (df_selected["Governance dimension"] == dimension) &
        (df_selected["Year"] == 2024)
    ][["Economy (name)", "Region", "Governance estimate (approx. -2.5 to +2.5)"]].copy()
    
    # Добавим отдельный "All Countries" для объединенного boxplot
    df_plot_all = df_plot.copy()
    df_plot_all["Region"] = "All Countries"
    
    # Объединяем с исходными регионами
    df_box = pd.concat([df_plot, df_plot_all], ignore_index=True)
    
    # Строим boxplot с hover, показывающим страну
    fig = px.box(
        df_box,
        x="Region",
        y="Governance estimate (approx. -2.5 to +2.5)",
        points="all",  # точки для всех стран
        color="Region",
        hover_data={
            "Economy (name)": True,  # выводим название страны
            "Region": True,
            "Governance estimate (approx. -2.5 to +2.5)": ':.2f'
        },
        title=f"WGI Estimate (2024) by Region ({dimension})",
        labels={"Governance estimate (approx. -2.5 to +2.5)": "Governance estimate (-2.5 to +2.5)"},
        template="plotly_white"
    )
    
    fig.update_layout(showlegend=False)
    fig.show()

# Интерактивный вывод
widgets.interactive(plot_wgi_boxplot, dimension=dimension_dropdown)
```
